## 1) Cargar el dataset sintético

## 3) Imputación de valores nulos

Probamos varias estrategias de imputación:
- Media (para variables numéricas)
- Mediana
- Moda
- KNNImputer (más avanzado)


```python
from sklearn.impute import SimpleImputer, KNNImputer
```

```python
df_copy = df.copy()
```

```python
# Imputación con media
imputer_mean = SimpleImputer(strategy="mean")
df_copy["age_imputed"] = imputer_mean.fit_transform(df_copy[["age"]])

df_copy[df_copy['age'].isna()].head()
```

```python
# Inputación con media a mano
df_copy = df.copy()
mean_age = df['age'].mean()
df_copy['age_imputed_manual'] = df_copy['age'].fillna(mean_age)
df_copy[df_copy['age'].isna()].head()
```

```python
# Imputación con mediana
df_copy = df.copy()
imputer_median = SimpleImputer(strategy="median")
df_copy["fare_imputed"] = imputer_median.fit_transform(df_copy[["fare"]])

df_copy[df_copy["fare"].isna()].head()
```

```python
# Imputación con moda
df_copy = df.copy()
imputer_most_frequent = df_copy['class'].mode()[0]
df_copy["class_imputed"] = df_copy["class"].fillna(imputer_most_frequent)

df_copy[df_copy["class"].isna()].head()
```

```python
# KNNImputer en conjunto numérico
num_cols = ["age", "fare"]
knn_imputer = KNNImputer(n_neighbors=5)
df_knn = pd.DataFrame(knn_imputer.fit_transform(df[num_cols]), columns=num_cols)
df_knn.head()
```

## 4) Codificación de variables categóricas

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

# OneHotEncoder en columna 'sex'
encoder = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
sex_encoded = encoder.fit_transform(df[['sex']])
sex_encoded[:5]
```

```python
# OrdinalEncoder en columna 'class'
ord_enc = OrdinalEncoder()
df['class_encoded'] = ord_enc.fit_transform(df[['class']])
df[['class', 'class_encoded']].head()
```

## 5) Escalado y normalización

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

scaler_std = StandardScaler()
scaler_minmax = MinMaxScaler()
scaler_robust = RobustScaler()

imputer_mean = SimpleImputer(strategy="mean")
df["age_imputed"] = imputer_mean.fit_transform(df[["age"]])

age_std = scaler_std.fit_transform(df[['age_imputed']])
age_minmax = scaler_minmax.fit_transform(df[['age_imputed']])
age_robust = scaler_robust.fit_transform(df[['age_imputed']])

# Para incluirlo en el DataFrame
# df['age_std'] = age_std
# df['age_minmax'] = age_minmax
# df['age_robust'] = age_robust

import matplotlib.pyplot as plt
plt.figure(figsize=(10,4))
plt.hist(age_std, bins=20, alpha=0.5, label='StandardScaler')
plt.hist(age_minmax, bins=20, alpha=0.5, label='MinMaxScaler')
plt.hist(age_robust, bins=20, alpha=0.5, label='RobustScaler')
plt.legend()
plt.title('Distribución de Edad con diferentes escalados')
plt.show()
```

## Normalización después del escalado

La normalización (por ejemplo L2) ajusta cada fila (cada muestra) para que la suma cuadrática de sus valores sea 1. A diferencia del *escalado* (que transforma cada columna), la normalización actúa sobre las filas. Es útil cuando:

- Comparamos muestras que tienen magnitudes globales distintas.
- Modelos basados en distancias (p.ej. KNN) pueden beneficiarse de que todas las muestras estén en la misma escala direccional.

A continuación aplicamos normalización L2 a dos columnas numéricas ya imputadas/escaladas. Si solo tenemos una columna, la normalización no aporta información (el valor se convertiría en 1).

```python
# Normalización L2 sobre las columnas numéricas imputadas
from sklearn.preprocessing import Normalizer
# Aseguramos imputación para 'fare' si tiene nulos
df_copy_norm = df.copy()
imputer_fare = SimpleImputer(strategy='median')
df_copy_norm['fare_imputed'] = imputer_fare.fit_transform(df_copy_norm[['fare']])

# Usamos 'age_imputed' (creada antes) y la nueva 'fare_imputed'
num_for_norm = df_copy_norm[['age_imputed', 'fare_imputed']].fillna(0)

normalizer = Normalizer(norm='l2')
num_norm = normalizer.fit_transform(num_for_norm)

# Opcional: añadir al DataFrame
# df_copy_norm['age_norm_l2'] = num_norm[:,0]
# df_copy_norm['fare_norm_l2'] = num_norm[:,1]

# Mostrar primeras filas normalizadas
pd.DataFrame(num_norm, columns=['age_norm_l2','fare_norm_l2']).head()
```

```python
# Gráficas comparando antes vs después de normalizar
import numpy as np
import matplotlib.pyplot as plt

# Recuperar datos usados en la normalización
age_vals = num_for_norm['age_imputed'].to_numpy().reshape(-1, 1)
fare_vals = num_for_norm['fare_imputed'].to_numpy().reshape(-1, 1)

# Componentes normalizados ya calculados en num_norm
age_norm_component = num_norm[:, 0]
fare_norm_component = num_norm[:, 1]

fig, axes = plt.subplots(1, 2, figsize=(12,4))

# Histograma: componente edad original (estandarizada) vs componente normalizada
axes[0].hist(age_vals.flatten(), bins=20, alpha=0.6, label='Edad (imputada)', color='tab:blue')
axes[0].hist(age_norm_component, bins=20, alpha=0.6, label='Edad normalizada (componente)', color='tab:orange')
axes[0].set_title('Histograma componente edad')
axes[0].legend()
axes[0].set_xlabel('Valor')
axes[0].set_ylabel('Frecuencia')

# Dispersión: antes vs después (en 2D)
axes[1].scatter(age_vals, fare_vals, alpha=0.5, label='Original', color='tab:green')
axes[1].scatter(age_norm_component, fare_norm_component, alpha=0.5, label='Normalizado (L2)', color='tab:red')
axes[1].set_title('Vectores antes y después de normalizar')
axes[1].set_xlabel('Edad / Edad normalizada')
axes[1].set_ylabel('Fare / Fare normalizado')
axes[1].legend()

plt.tight_layout()
plt.show()

# (Opcional) Verificar que cada vector normalizado tiene norma ~1
normas = np.sqrt(age_norm_component**2 + fare_norm_component**2)
print('Norma media:', normas.mean())
print('Rango de normas:', normas.min(), '->', normas.max())
```
